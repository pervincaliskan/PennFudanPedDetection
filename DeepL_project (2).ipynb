```python
import os
import numpy as np
import torch
from PIL import Image
import torchvision
from torchvision.models.detection import fasterrcnn_resnet50_fpn
from torchvision.transforms import functional as F
import matplotlib.pyplot as plt
import random
```

Veri Setini İndirme ve Hazırlık İçin Dataset Sınıfını Ekleme

```python
class PennFudanDataset(torch.utils.data.Dataset):
    def __init__(self, root, transforms=None):
        self.root = root
        self.transforms = transforms # Uygulanacak dönüşümler
        self.imgs = list(sorted(os.listdir(os.path.join(root, "PNGImages"))))
        self.masks = list(sorted(os.listdir(os.path.join(root, "PedMasks"))))

    def __getitem__(self, idx):
        img_path = os.path.join(self.root, "PNGImages", self.imgs[idx])
        mask_path = os.path.join(self.root, "PedMasks", self.masks[idx])
        img = Image.open(img_path).convert("RGB") # Görüntüyü RGB formatında aç
        mask = Image.open(mask_path)
# Maske, görüntüdeki her pikselin hangi nesneye ait olduğunu gösteren bir etiketleme görüntüsüdür
        mask = np.array(mask) # Maskeyi numpy dizisine dönüştür
        obj_ids = np.unique(mask)[1:]  # 0 arka plan
        masks = mask == obj_ids[:, None, None] # Her nesne için binary maske oluştur


        boxes = []
        for m in masks:
            pos = np.where(m) # Maskede True olan piksellerin yani nesneye ait olan koordinatlarını bul
            xmin = np.min(pos[1])
            xmax = np.max(pos[1])
            ymin = np.min(pos[0])
            ymax = np.max(pos[0])
            boxes.append([xmin, ymin, xmax, ymax]) # Sınırlayıcı kutuyu ekle

        boxes = torch.as_tensor(boxes, dtype=torch.float32) # Kutuları tensöre dönüştür
        labels = torch.ones((len(obj_ids),), dtype=torch.int64)  # sadece kişi sınıfı
        masks = torch.as_tensor(masks, dtype=torch.uint8) # Maskeleri tensöre dönüştür

        image_id = torch.tensor([idx])  # Görüntü kimliği
        area = (boxes[:, 3] - boxes[:, 1]) * (boxes[:, 2] - boxes[:, 0]) # Her sınırlayıcı kutunun alanı (piksel cinsinden)
        iscrowd = torch.zeros((len(obj_ids),), dtype=torch.int64)

        target = {
            "boxes": boxes,
            "labels": labels,
            "masks": masks,
            "image_id": image_id,
            "area": area,
            "iscrowd": iscrowd
        } # Tüm hedef verileri bir sözlükte topla

        if self.transforms:
            img = self.transforms(img) # Dönüşümleri uygula

        return img, target # Görüntü ve hedef sözlüğünü döndür

    def __len__(self):
        return len(self.imgs)  # Veri setindeki toplam görüntü sayısını döndür
```

 Veri Setini Bölme ve Veri Yükleyicileri Oluşturma

```python
from torch.utils.data import random_split, DataLoader
import torchvision.transforms as T

transform = T.Compose([T.ToTensor()]) # Görüntüyü tensöre dönüştürme işlemi,Görüntüyü modelin beklediği formata getirdik, [0-255] aralığından [0-1] aralığına normalize eder
dataset = PennFudanDataset("PennFudanPed", transforms=transform) # Veri setini oluştur

# %60 train, %20 val, %20 test
total = len(dataset)
train_size = int(0.6 * total)
val_size = int(0.2 * total)
test_size = total - train_size - val_size

train_ds, val_ds, test_ds = random_split(dataset, [train_size, val_size, test_size])

train_loader = DataLoader(train_ds, batch_size=2, shuffle=True, collate_fn=lambda x: tuple(zip(*x)))
val_loader = DataLoader(val_ds, batch_size=2, shuffle=False, collate_fn=lambda x: tuple(zip(*x)))
test_loader = DataLoader(test_ds, batch_size=1, shuffle=False, collate_fn=lambda x: tuple(zip(*x)))
```

Model Tanımlama

```python
def get_model(num_classes):
    model = fasterrcnn_resnet50_fpn(pretrained=True)
    in_features = model.roi_heads.box_predictor.cls_score.in_features # Giriş özelliklerinin sayısını al
    model.roi_heads.box_predictor = torchvision.models.detection.faster_rcnn.FastRCNNPredictor(in_features, num_classes) # Sınıflandırıcıyı değiştir
    return model
# Cihazı belirle (GPU veya CPU)
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model = get_model(num_classes=2)  # 2 sınıflı model oluşturduk (1: insan, 0: arka plan)
model.to(device)
```

Optimizasyon ,Eğitim ve Doğrulama

```python
import torch.optim as optim

# Optimizasyon algoritmasını tanımla (SGD)
params = [p for p in model.parameters() if p.requires_grad] # Eğitilebilir parametreleri seç
optimizer = optim.SGD(params, lr=0.005, momentum=0.9, weight_decay=0.0005) # SGD optimizasyonu tanımla
# Öğrenme oranı çizelgeleyicisini tanımla (StepLR)
lr_scheduler = optim.lr_scheduler.StepLR(optimizer, step_size=5, gamma=0.1)

# Tek bir epok için eğitim fonksiyonu
def train_one_epoch(model, optimizer, loader, device):
    model.train()
    total_loss = 0
    for imgs, targets in loader: # Veri yükleyiciden batch'leri al
        imgs = list(img.to(device) for img in imgs) # Görüntüleri cihaza taşı
        targets = [{k: v.to(device) for k, v in t.items()} for t in targets]
        loss_dict = model(imgs, targets) # İleri geçiş ve kayıp hesaplama
        losses = sum(loss for loss in loss_dict.values()) # Tüm kayıpları topla
        optimizer.zero_grad()
        losses.backward()  # Geri yayılım yapıp model parametrelerini günceller
        optimizer.step()
        total_loss += losses.item()
    return total_loss / len(loader)

# Tek bir epok için doğrulama fonksiyonu
def validate_one_epoch(model, loader, device):
    # Modeli değerlendirme moduna almadan önce mevcut durumunu sakla
    original_mode = model.training
    # Geçici olarak training moduna al. Faster R-CNN'nin kayıp hesaplaması için modelin eğitim modunda olması gerekir.
    model.train()
    total_loss = 0
    with torch.no_grad():  # Gradyan hesaplamayı devre dışı bırak. Doğrulama sırasında parametre güncellemesi yapmayacağımız
        for imgs, targets in loader:
            imgs = list(img.to(device) for img in imgs)
            targets = [{k: v.to(device) for k, v in t.items()} for t in targets]
            # Training modunda olduğu için loss dictionary döndürecek
            loss_dict = model(imgs, targets)
            losses = sum(loss for loss in loss_dict.values())
            total_loss += losses.item()
    # Orijinal moda geri dön
    model.train(original_mode)
    return total_loss / len(loader) # Ortalama kaybı döndür
```

Eğitim

```python

num_epochs = 10
train_losses = []
val_losses = [] # kayıpları saklamak için liste

for epoch in range(num_epochs):
    # Eğitim aşaması
    train_loss = train_one_epoch(model, optimizer, train_loader, device)
    train_losses.append(train_loss)

    # Doğrulama aşaması
    val_loss = validate_one_epoch(model, val_loader, device)
    val_losses.append(val_loss)

    lr_scheduler.step() # Öğrenme oranını güncelle
    print(f"Epoch [{epoch+1}/{num_epochs}] Train Loss: {train_loss:.4f}, Val Loss: {val_loss:.4f}")
```

```python

# Eğitim ve Doğrulama Kayıp Grafiği
plt.plot(train_losses, label='Eğitim Kaybı')
plt.plot(val_losses, label='Doğrulama Kaybı')
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.title("Eğitim ve Doğrulama Kayıp Grafiği")
plt.legend() # Etiketleri göstermek için lejant ekle
plt.show()
```

Test Seti Üzerinde Değerlendirme ve Görselleştirme

```python
import random

model.eval() # Modeli değerlendirme modu
with torch.no_grad(): # Gradyan hesaplamayı devre dışı bırak
    # test_loader'ı listeye çevirip
    test_samples_list = list(test_loader)
    num_test_samples = len(test_samples_list)
    num_samples_to_show = min(10, num_test_samples)

    selected_indices = random.sample(range(num_test_samples), num_samples_to_show) #random örnekler alma

    for idx in selected_indices: # Seçilen indeksler üzerinde döngü
        imgs, targets = test_samples_list[idx]  # Görüntü ve hedef al
        img = imgs[0].to(device) # Batch_size=1 olduğu için, her batch'te sadece 1 görüntü var
        output = model([img])[0] # Modelden tahmin al

        img = img.permute(1, 2, 0).cpu().numpy() # Görüntüyü NumPy dizisine dönüştür Tensörün boyutlarını yeniden düzenler
        plt.imshow(img) # Görüntüyü göster

        for i, box in enumerate(output['boxes']):  # Tahmin edilen her kutu için
            box = box.cpu().numpy()  # Kutuyu CPU'ya taşı ve NumPy'a dönüştür
            xmin, ymin, xmax, ymax = box  # Kutu koordinatlarını al
            label = output['labels'][i].item()  # Sınıf etiketini al
            # Sadece kişi sınıfını ve belirli bir güven skorunun üzerindekileri gösterelim
            score = output['scores'][i].item()  # Güven skorunu al
            if label == 1 and score > 0.5:  # Güven skoru 0.5'ten büyükse göster
                label_name = "person"  # Sınıf adı

                plt.gca().add_patch( # Güncel Eksen nesnesini döndürür.
                    plt.Rectangle((xmin, ymin), xmax - xmin, ymax - ymin,
                                  edgecolor='red', facecolor='none', linewidth=2)) # Kutuyu çiz
                plt.text(xmin, ymin, f"{label_name}: {score:.2f}", color='white',
                         fontsize=8, bbox=dict(facecolor='red', alpha=0.5)) # Etiket ve skoru yaz

        plt.axis("off") # Eksenleri gizle
        plt.title("Tahmin: person")
        plt.show()
```
